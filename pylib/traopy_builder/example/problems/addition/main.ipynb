````python
import traopy_builder as trab
from pathlib import Path
````

````python
builder = trab.ProcedureBuilder()
````

````python
base_path = Path(".").resolve()
input_text = trab.ResourceKind.TextFile(
    trab.Text(
        "input",
        f"{base_path}/testcases/input.txt",
    )
)
output_text = trab.ResourceKind.TextFile(
    trab.Text(
        "output",
        f"{base_path}/testcases/output.txt",
    )
)

input_result = builder.add_resource(input_text)
output_result = builder.add_resource(output_text)
````

````python
actual_output = trab.ResourceKind.RuntimeTextFile(
    trab.PyRuntimeText(
        "actual",
        "actual"
    )
)

actual_result = builder.add_resource(actual_output)
````

````python
script = trab.Text(
    "script",
    f"{base_path}/tester.py",
)

script_result = builder.add_script(script)
````

````python
execution = trab.Execution(
    "test",
    script_result,
    [
        trab.PyDependency(
            input_result,
            "INPUT",
        ),
        trab.PyDependency(
            output_result,
            "OUTPUT",
        ),
    ],
)

execution_result = builder.add_execution(execution)
````

````python
builder.write_to("judge_procedure.json")
````

````python
local_judge = traopy.LocalJudge(base_path.joinpath("tempdir"))
result = await local_judge.run(builder, {"actual": "3\n"})
import json
parsed = json.loads(result)
pretty = json.dumps(parsed, indent=4)
print(pretty)
````

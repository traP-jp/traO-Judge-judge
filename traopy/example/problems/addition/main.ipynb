````python
import traopy
from pathlib import Path
````

````python
judge_env = traopy.Environment()
````

````python
base_path = Path(".").resolve()
input_text = traopy.PyResourceKind.TextFile(
    traopy.PyText(
        "input",
        f"{base_path}/testcases/input.txt",
    )
)
output_text = traopy.PyResourceKind.TextFile(
    traopy.PyText(
        "output",
        f"{base_path}/testcases/output.txt",
    )
)

input_result = judge_env.add_resource(input_text)
output_result = judge_env.add_resource(output_text)
````

````python
actual_output = traopy.PyResourceKind.OnetimeTextFile(
    traopy.PyOnetimeText(
        "actual",
    )
)

actual_result = judge_env.add_resource(actual_output)
````

````python
script = traopy.PyScript(
    "script",
    f"{base_path}/tester.py",
)

script_result = judge_env.add_script(script)
````

````python
execution = traopy.PyExecution(
    "test",
    script_result,
    [
        traopy.PyDependency(
            input_result,
            "INPUT",
        ),
        traopy.PyDependency(
            output_result,
            "OUTPUT",
        ),
        traopy.PyDependency(
            actual_result,
            "ACTUAL",
        ),
    ],
)

execution_result = judge_env.add_execution(execution)
````

````python
judge_env.write_to("judge_procedure.json")
````
